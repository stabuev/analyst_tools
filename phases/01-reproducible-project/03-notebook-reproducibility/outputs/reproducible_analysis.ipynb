# Воспроизводимый расчёт среднего чека

Notebook хранится без outputs и выполняется строго сверху вниз. Данные встроены только для учебного примера.

```python
import csv
import io
import statistics
```

```python
RAW_CSV = """order_id,amount
A-001,120.0
A-002,80.0
A-003,100.0
"""

rows = list(csv.DictReader(io.StringIO(RAW_CSV)))
assert rows, "Входные данные пусты"
```

```python
def summarize_orders(records):
    amounts = [float(record["amount"]) for record in records]
    return {
        "orders": len(amounts),
        "revenue": sum(amounts),
        "average_order_value": statistics.mean(amounts),
    }

summary = summarize_orders(rows)
assert summary == {
    "orders": 3,
    "revenue": 300.0,
    "average_order_value": 100.0,
}
summary
```

## Вывод

Контрольный набор содержит 3 заказа, выручку 300 и средний чек 100. Assert связывает бизнес-вывод с вычисленным результатом.
